**Transmitter Codes:**


```python
#include <SPI.h> // SPI communication with nRF24L01+.
#include <RF24.h> // nRF24L01+ library v1.4.1
#include <Wire.h> // I2C communication with Mpu6050.
#include <MPU6050.h> //Mpu6050 library v0.3.0

MPU6050 acceleration sensor; // create object named acceleration_sensor.
int x, y, z; // acceleration movements of x, y, z planes.

RF24 radio(7, 8); // create object named radio. CE and CSN pins.
int data[2]; // Define array for x and y plane.

invalid setup() {
    Wire.begin(); // I2C communication is started.
    acceleration_sensor.initialize(); // Initialize Mpu6050.

    radio.begin(); // initialize nRF24L01+.
    radio.openWritingPipe(1234); // Communication key 1234.
}

void loop() {
    acceleration_sensor.getAcceleration(&x, &y, &z); // Get acceleration from the sensor.

    data[0] = map(x, -17000, 17000, -255, 255 ); //data of the X plane (forward-backward)
    data[1] = map(y, -17000, 17000, -255, 255); // Data of the Y plane (right-left)
    radio.write(data, sizeof(data)); // send information from data variable to nRF24L01+.
}
```

**Receiver Codes:**

```python
#include <SPI.h> // SPI communication with nRF24L01+.

#include <RF24.h> // nRF24L01+ library v1.4.1
 

RF24 radio(7, 8); // create object named radio. CE and CSN pins.

int data[2];// Define array for X and Y plane
const byte IN1 = 5, IN2 = 6, IN3 = 9, IN4 = 10; // PWM pins. (jumps installed on enA and enB).

void setup() {
  radio.begin(); // initialize nRF24L01+.
  radio.openReadingPipe(0, 1234); // Communication key 1234.
  radio.startListening(); // Switch to receiver mode.
}

void loop() {
  if (radio.available()) { // If data is coming
    radio.read(data, sizeof(data)); // Load the incoming data into the data variable.

    if (data[0] > 50) { //Next.
      analogWrite(IN1, data[0]);
      analogWrite(IN2, 0);
      analogWrite(IN3, data[0]);
      analogWrite(IN4, 0);
    }
    if (data[0] < 50) { //Back.
      analogWrite(IN1, 0);
      analogWrite(IN2, data[0]);
      analogWrite(IN3, 0);
      analogWrite(IN4, data[0]);
    }
    if (data[1] > 50) { //Left.
      analogWrite(IN1, 0);
      analogWrite(IN2, data[1]);
      analogWrite(IN3, data[1]);
      analogWrite(IN4, 0);
    }
    if (data[1] < 50) { // Right.
      analogWrite(IN1, data[1]);
      analogWrite(IN2, 0);
      analogWrite(IN3, 0);
      analogWrite(IN4, data[1]);
    }
    if (data[0] > 50 && data[0] < 50 && data[1] > 50 && data[1] < 50) { // Stop.
      analogWrite(IN1, 0);
      analogWrite(IN2, 0);
      analogWrite(IN3, 0);
      analogWrite(IN4, 0);
    }
  }
}
```
